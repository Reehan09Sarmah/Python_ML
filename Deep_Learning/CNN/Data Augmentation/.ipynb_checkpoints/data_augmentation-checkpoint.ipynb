```python
from keras.preprocessing import image
from keras.preprocessing.image import ImageDataGenerator
```

```python
img = image.load_img('train/dog.jpg', target_size = (200, 200))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(img)
```

```python
type(img)
```

```python
datagen = ImageDataGenerator(
    rotation_range=30, # angle to rotate the image by
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    width_shift_range=0.2,
    height_shift_range=0.2,
    # fill_mode='constant'->as pixels shift, empty places becomes black
    #by default ->fill_mode='nearest'->as pixels shift, empty places take value of nearest pixel
    
)
```

```python
img = image.img_to_array(img) # convert it to numpy array
```

```python
img.shape
```

```python
input_batch = img.reshape(1, 200, 200, 3) # 1 image of 200x200x3 dimension
```

```python
i=0

# loop to go on creating new images from 1 image.
# flow is used when working on 1 image.
for output in datagen.flow(input_batch, batch_size=1, save_to_dir='aug'):
    i = i + 1
    if i == 10:
        break
```
