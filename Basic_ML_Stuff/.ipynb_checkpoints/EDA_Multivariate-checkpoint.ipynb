```python
import pandas as pd
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
tips = sns.load_dataset('tips')
titanic = pd.read_csv('content/train.csv')
flights = sns.load_dataset('flights')
iris = sns.load_dataset('iris')
```

### 1. Scatterplot (Numerical - Numerical)

```python
tips.head()
```

```python
sns.scatterplot(tips, x='total_bill', y='tip', hue='sex', style='smoker', size='size')
# plotting total bill they ate vs the tip they gave.
# We have separated the male vs female through color
# we have separated smoker and non_smoker by point shape.
# size of each datapoint depends on its corresponding size value
```

### 2. Barplot (Numerical - Categorical)

```python
titanic.head()
```

```python
sns.barplot(titanic, x='Pclass', y='Age', hue='Sex')
```

We can see that with the change in class, the average age differs. Class 1 has older people, 2 has slightly younger and 1 younger. Average age of male is greater than that of female in each class

```python
sns.barplot(titanic,x='Pclass', y='Fare', hue='Sex')
```

#### Boxplot (Numerical - Categorical)

```python
sns.boxplot(titanic, x='Sex',y='Age', hue='Survived')
```

In case of male, Younger population has more survival
In case of Female, Older than male survivors
