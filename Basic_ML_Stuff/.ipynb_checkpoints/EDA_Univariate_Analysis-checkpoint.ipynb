## Univariate Analysis
##### A dataset contains multiple columns. Each column can be called a variable. Independent Analysis of such a variable is called Univariate Analysis.
Data has 2 types --> Numerical (height, age, etc) and Categorical (country, Course, etc). Let's do it on our titanic dataset. It is a dataset of passengers who were travelling in the titanic. Target variable is 0 or 1. If the passenger survived or not

```python
import pandas as pd
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('content/train.csv')
```

```python
df.head()
```

### 1. Categorical Data

#### a. Countplot
Count of categorical data

##### For Survived Column

```python
df['Survived'].value_counts()
```

```python
# df['Survived'].value_counts().plot(kind='bar')
```

```python
sns.countplot(data=df, x='Survived' )  # using seaborn
```

Here, less people survived as compared to the number of people who didn't

##### For pclass

```python
df['Pclass'].value_counts()
```

```python
sns.countplot(df, x='Pclass')
```

Maximum number of people travelled in class 3. Next in 1 and then in 2. It may also seem like that there was not much fare difference between 1 and 2 so some people might have gone for 1 instead of 2

##### Gender Distribution

```python
df['Sex'].value_counts()
```

```python
sns.countplot(df, x='Sex')
```

Here, There were more male passengers than female

##### For the station that the people embardked from

```python
df['Embarked'].value_counts()
```

```python
sns.countplot(df, x='Embarked')
```

Here, we can seem that maximum people embarked from station S, then C and then Q.

### b. Piechart

```python
# The distribution is in percentage
df['Survived'].value_counts().plot(kind='pie', autopct='%.2f',legend=True,title='Survived or Not?')
```

```python
df['Pclass'].value_counts().plot(kind='pie', autopct='%.2f',legend=True,title='Passenger Class')
```

```python
df['Sex'].value_counts().plot(kind='pie', autopct='%.2f',legend=True,title='Sex Percentage')
```

### 3. Numerical Data

#### a. Histogram

```python
import matplotlib.pyplot as plt
plt.hist(df['Age'])  
# distribution of data in numerical column. Change the bins to change the range
```

#### b. Distplot
Plots the Probability Distribution of the Ages. We call it the Probability Distribution Function (PDF). Y axis has the probability and X axis has the ages

```python
sns.distplot(df['Age']) 
```

#### c. Boxplot
To know about noisy data and outliers
Q1 -> Quartile 1 (25% data is below this value)
Median -> 50 percentile (50% data is below this)
Q3 -> Quartile 3 -> 75% data is below this
Leftmost part is the Min Value. (Q1 - 1.5 * IQR)
Rightmost part  is the Max value (Q3 + 1.5 * IQR)
Min Value: Most of the values in column are above it
Max value: Most of the values in column are below it

```python
sns.boxplot(df, x='Fare')
```

Here, Min and Max values fall between 0 and 100. Outlier are captured whose values are more than n. (0<n<100)

```python
sns.boxplot(df, x='Age')
```

Here, Median falls near to 30. Maximum number of ages are below this number and it is somewhere close to 65. Minimum Age is close to 0. Others are above it. More than 65 are outliers that are captured.
