```python
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
tips = sns.load_dataset("tips")
iris = sns.load_dataset("iris")
titanic = sns.load_dataset("titanic")
planets = sns.load_dataset("planets")
```

```python
tips
```

### Scatter Plot

```python
sns.scatterplot(x='tip', y='total_bill',data=tips, 
                hue='day', size="size", palette='YlGnBu')
# We made x axis = tip column, y axis = total_bill column
# we set hue to day column, set different colors to points based on their corresponding values in 'day' column
# we also set the size to size column. Different datapoints different size based on their corresponding 'size' value
# we set the palette to Yellow, Green and Blue. Changed the points color
```

### Histograms

```python
sns.histplot(tips['tip'], kde=True, bins=10)
# kde displays the distribution curve.
# bins stands for the number of vertical bars you want
```

##### OR similarly

```python
sns.displot(tips['tip'], kde=True, bins=10)
```

### Barplots

```python
sns.barplot(x='sex', y='tip', data=tips, palette='GnBu')
```

### Boxplot
Consists of the minimum, maximum, median value. And also the outliers (the black dots)

```python
sns.boxplot(x='day', y='tip', data=tips, hue='sex', palette='YlGnBu')
```

### Striplot

```python
sns.stripplot(x='day', y='tip', data=tips, hue='sex', palette='YlGnBu', dodge=True)
# if dodge = False then the males and females will be mixed.
# Now they are separately shown
```

### Joint Plots

```python
sns.jointplot(x='tip', y='total_bill', data=tips)
```

```python
sns.jointplot(x='tip', y='total_bill', data=tips, kind='hex',cmap='YlGnBu')
```

### Pairplots

```python
sns.pairplot(titanic.select_dtypes(['number']), hue='pclass')
```

### Correlations are Important
##### Use Heatmap

```python
titanic.corr(numeric_only=True)
```

```python
sns.heatmap(titanic.corr(numeric_only=True), annot=True, cmap='YlGnBu')
```

```python
sns.heatmap(titanic.corr(numeric_only=True), annot=True, cmap='coolwarm')
```
